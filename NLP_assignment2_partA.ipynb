## Majority Baseline

```python
# determine the majority class based on the training data
def majority_class(train_labels):
    num_of_OFF = 0
    num_of_NOT = 0
    for j in train_labels:
        if j == 1:
            num_of_OFF += 1
        else:
            num_of_NOT += 1
    if num_of_OFF > num_of_NOT:
        majority_class = 1
    else:
        majority_class = 0
    return majority_class

#example on training data
print(majority_class(df.labels))
```

```python
y_pred = [majority_class(df.labels) for i in range(len(test_set))]
#y_pred
```

```python
y_true = test_set['labels'].tolist()
#y_true
```

```python
print(classification_report(y_true, y_pred))
```

## Random Baseline

```python
import random

def random_item():
    list_choice = [1, 0]# choices
    random_item = random.choice(list_choice)# Chooses from list_choice
    return(random_item)

print(random_item())
```

```python
y_pred = [random_item() for i in range(len(test_set))]
y_pred
```

```python
y_true = test_set['labels'].tolist()
y_true
```

```python
print(classification_report(y_true, y_pred))
```

##### But running randomized one time is not feabible. Therefore, average out over 100 iterations

```python
random.seed(4)

# Average scores out over 100 iterations
n = 100

precision_scores_OFF = []
precision_scores_N = []
recall_scores_OFF = []
recall_scores_N = []
f1_scores_OFF = []
f1_scores_N = []

weighted_avg_prec = []
weighted_avg_rec = []
weighted_avg_f1 = []

macro_avg_prec = []
macro_avg_rec = []
macro_avg_f1 = []

for i in range(n):
    
    y_pred = [random_item() for i in range(len(test_set))]
    
    precision_scores_N.append(precision_recall_fscore_support(y_true, y_pred, average=None)[0][0])
    precision_scores_OFF.append(precision_recall_fscore_support(y_true, y_pred, average=None)[0][1])
    recall_scores_N.append(precision_recall_fscore_support(y_true, y_pred, average=None)[1][0])
    recall_scores_OFF.append(precision_recall_fscore_support(y_true, y_pred, average=None)[1][1])
    f1_scores_N.append(precision_recall_fscore_support(y_true, y_pred, average=None)[2][0])
    f1_scores_OFF.append(precision_recall_fscore_support(y_true, y_pred, average=None)[2][1])
    weighted_avg_prec.append(precision_recall_fscore_support(y_true, y_pred, average='weighted')[0])
    weighted_avg_rec.append(precision_recall_fscore_support(y_true, y_pred, average='weighted')[1])
    weighted_avg_f1.append(precision_recall_fscore_support(y_true, y_pred, average='weighted')[2])
    macro_avg_prec.append(precision_recall_fscore_support(y_true, y_pred, average='macro')[0])
    macro_avg_rec.append(precision_recall_fscore_support(y_true, y_pred, average='macro')[1])
    macro_avg_f1.append(precision_recall_fscore_support(y_true, y_pred, average='macro')[2])


print('Precision_OFF over 100 iterations is:',np.mean(precision_scores_OFF))
print('Precision_NOT over 100 iterations is:',np.mean(precision_scores_N)) 
print('Recall_OFF over 100 iterations is:',np.mean(recall_scores_OFF)) 
print('Recall_NOT over 100 iterations is:',np.mean(recall_scores_N))
print('f1_OFF over 100 iterations is:',np.mean(f1_scores_OFF))
print('f1_NOT over 100 iterations is:',np.mean(f1_scores_N))
print('weighted_avg_prec over 100 iterations is:',np.mean(weighted_avg_prec))
print('weighted_avg_rec over 100 iterations is:',np.mean(weighted_avg_rec))
print('weighted_avg_f1 over 100 iterations is:',np.mean(weighted_avg_f1))
print('macro_avg_prec over 100 iterations is:',np.mean(macro_avg_prec))
print('macro_avg_rec over 100 iterations is:',np.mean(macro_avg_rec))
print('macro_avg_f1 over 100 iterations is:',np.mean(macro_avg_f1))
```

#### Total number of subwords (Note: subwords are defined as tokens that have at least a subword (e.g. 'text' --> 0 subwords, 'URL' (U, ##RL) --> 2 subwords) (4b)


```python
# total number of subwords on small example (for testing purposes)

counter = 0
sentence = ['I', 'love', 'em', '##bed', '##ding', '##s', 'like', 'be', '##rt', 'oppo', '##eee', '##20', '##40','iii', '.']
sentence1 = ['I', 'love', 'em', '##bed', '##ding', '##s', 'like', 'be', '##rt']
sentence2 = ['You', 'hate', 'football', 'sub', '##word', '.']
sentence3 = ['He', 'likes', 'U', '##RL', '##s', 'be', '##rt', '.']

ssss = [sentence1, sentence2, sentence3]
ssss = df['text'][0:3]

for s in ssss:
  sentence = model.tokenizer.tokenize(s)
  print(sentence)

  for i in range(len(sentence)-1):
    c = 0
    if '##' not in sentence[i] and '##' in sentence[i+1]:
      print(sentence[i])
      counter += 1
      while c < len(sentence)-i-1:
        c+=1
        if '##' in sentence[i+c]:
          print(sentence[i+c])
          counter += 1
        else:
          #print('\n')
          break
      
print('\n')
print(counter)
```
